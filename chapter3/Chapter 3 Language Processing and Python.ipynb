# 3.4 REgular Expressions for Detecting Word Patterns

```python
import re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

### Use Basic Matacharacters

```python
# find words ending with ed using ed$

[w for w in wordlist if re.search('ed$', w)]
```

```python
# . symbel match any character: $ end ^ start

[w for w in wordlist if re.search('^..j..t..$', w)]
```

```python
# without ^ and $ there is no formal start and end length

[w for w in wordlist if re.search('..j..t..', w)]
```

```python
# optional ?

l = ['email', 'e-mail']
[w for w in l if re.search('^e-?mail$', w)]
```

### Range and Closure

```python
# T9: system is used for entering text on mobile phones
# Two or more words that are entered with the same sequence
# of keystrokes are known as textonyms.
# [] means or, the order does not matter [or]==[ro]

[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
```

```python
# + means one or more and * means zero or more
# also referred to as closure

[w for w in wordlist if re.search('^[ghi]+[mno]+[jlk]+[def]+$', w)]
```

```python
[w for w in wordlist if re.search('^[ghi][mno]+[jlk]+[def]*$', w)]
```

```python
# - means between 
[w for w in wordlist if re.search('^[g-i][m-o]+[jlk]+[def]*$', w)]
```

```python
# - means between 
[w for w in wordlist if re.search(r'^[g-i][m-o]+[jlk]+[def]*$', w)]
```

    ^ is different when in bracket, means not
    \ can use special characters \. == .
    {m} specify the number of matches
    A|B either A or B
    () define the scope
    {m, } at least m
    {, m} at most m
    {m, n} at least m at most n

### Doing More with Word Pieces

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'

def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')

print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]

cfd = nltk.ConditionalFreqDist(cvs)
cfd.tabulate()
```

```python
cv_word_pairs = [(cv, w) for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]

cv_index = nltk.Index(cv_word_pairs)

cv_index['su']

cv_index['po']
```

### Finding Word Stems

```python
def stem(word):
    for suffix in ['ing', 'ly', 'ed', 'ious', 'ies', 'ive', 'es', 's', 'ment']:
        if word.endswith(suffix):
            return word[:-len(suffix)]
        return word
```

```python
re.findall(r'^.*(ing|ly|ed|ious|ies|ive|es|s|ment)$', 'processing')
```

```python
# to match does not contain, we use ?:

re.findall(r'^.*(?:ing|ly|ed|ious|ies|ive|es|s|ment)$', 'processing')
```

### Stemmers: find the word root

```python
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()

print([porter.stem(t) for t in tokens])
print([lancaster.stem(t) for t in tokens])
```

```python
class IndexedText(object):
    
    def __init__(self, stemmer, text):
        self._text = text
        self._stemmer = stemmer
        self._index = nltk.Index((self._stem(word), i) 
                                 for (i, word) in enumerate(text))
    
    def concordance(self, word, width=40):
        key = self._stem(word)
        wc = width//4
        for i in self._index[key]:
            lcontext = ' '.join(self._text[i-wc:i])
            rcontext = ' '.join(self._text[i:i+wc])
            ldisplay = '%*s' %(width, lcontext[-width:])
            rdisplay = '%-*s' %(width, rcontext[:width])
            print(f'{ldisplay} {rdisplay}')
    
    def _stem(self, word):
        return self._stemmer.stem(word).lower()
```

```python
porter = nltk.PorterStemmer()
grail = nltk.corpus.webtext.words('grail.txt')
text = IndexedText(porter, grail)
text.concordance('lie')
```

### NLTK Regular Expression Tokenizer

```python
text = 'That U.S.A. poster-print costs $12.40...'

pattern = r'''(?x) # set flag to allow verbose regexps
    ([A-Z]\.)+ # abbreviations, e.g. U.S.A.
    | \w+(-\w+)* # words with optional internal hyphens
    | \$?\d+(\.\d+)?%? # currency and percentages, e.g. $12.40, 82%
    | \.\.\. # ellipsis
    | [][.,;"'?():-_`] # these are separate tokens
'''

nltk.regexp_tokenize(text, pattern)
```

# 3.8 Segmentation

    Punkt sentence segmenter 
    Word Segmentation
    

```python
sent_tokenizer=nltk.data.load('tokenizers/punkt/english.pickle')
text = nltk.corpus.gutenberg.raw('chesterton-thursday.txt')
sents = sent_tokenizer.tokenize(text)

pprint.pprint(sents[171:181])
```

```python
def segment(text, segs):
    words = []
    last = 0
    for i in range(len(segs)):
        if segs[i] == '1':
            words.append(text[last:i+1])
            last = i+1
    words.append(text[last:])
    return words
```

```python
text = "doyouseethekittyseethedoggydoyoulikethekittylikethedoggy"
seg1 = "0000000000000001000000000010000000000000000100000000000"
seg2 = "0100100100100001001001000010100100010010000100010010000"
```

```python
segment(text, seg1)
```

```python
segment(text, seg2)
```

    We assume the learner is acquiring words and storing them in an internal lexicon. Given a suitable lexicon, it is possible to reconstruct the source text as a sequence of lexical items.
